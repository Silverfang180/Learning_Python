```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Employees.csv')# retrive dataset
df.head(5)
```

```python
# Which Country Emplpoyees has more rating?
# Solution 1 with detailed info.
df[(df['Rating']>7)&df['Country']]
```

```python
#Solution 2 which is correct way to represent
df.groupby('Country')['Rating'].mean().sort_values()
```

```python
# Country which has highest employees rating
df.groupby('Country')['Rating'].mean().sort_values(ascending=False)[:1]
```

```python
# If we want to visualize the data
df.groupby('Country')['Rating'].mean().sort_values(ascending=False).plot(kind='pie')
```

```python
# If we want to visualize the data
df.groupby('Country')['Rating'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
# List 3 states based on rating
df.groupby('State')['Rating'].mean().sort_values(ascending=True)[:3]
```

```python
df.groupby('State')['Rating'].mean().sort_values(ascending=False).plot(kind='hist')
```

```python
# Number of employees who has loan them?
df['Loan'].value_counts().plot(kind='bar')
plt.title('Number of Loans')
plt.xlabel('Loan')
plt.ylabel('Employees Counts')
```
