**SETS**
- set is an unordered collection of heterogeneous data types values.
- Set doesn't support indexing and slicing
- Set is mutable in nature. But, we can't modify the element and we can add new element or remove existing element.
- Set doesn't allow duplicate elements.
- Set element enclosed in {} and separated by commas --> **,**

- In sets TRUE --> 1, FALSE --> 0

```python
set1={1,3,'choma', 'faizal', None, 'zeros', 0, False}
set1
```

```python
#Addition
set1={1,3,'choma', 'faizal', None, 'zeros', 0, False}
set1.add(5)
set1
```

```python
#Update:- in update operation multiple elements can be added into set

set1.update('Abhishke','faizal')
set1
```

```python
# Remove

set1.remove('choma')
```

```python
set1
```

```python
# Discard
set1.discard('faizal')
```

```python
set1
```

# **Dictionary**
- Dictionary is an ordered collections of Key-value pairs structures .
- Dictionary doesn't allow indexing and slicing. But, here key is used as indexing  .
- it is mutable in nature .
- it doesn't allow duplicate keys .
- it is enclosed in {} and in key-value pairs .

```python
report={"Name":"Rocky bhai", "greetings":"salam rocky bhai"}
print(report)
```

```python
report['Name']
```

```python
report = {'name':["Faizal", 'Abhisehek','Nigga'], 'Marks':[90,90,90,]}# IN single key we ca add multiple values with the help of list
report
```

```python
report = {'name':("Faizal", 'Abhisehek','Nigga'), 'Marks':(90,90,90,)}# To retrive an perticular element from list in dictionary we specify index
report["name"][1], report['Marks'][2]
```

```python
report.update({'name':['Abhi','Abhishek','Faizal']})
report
```

```python
report = {'name':("Faizal", 'Abhisehek','Nigga'), 'Marks':(90,90,90,)}
report.pop('name')
report
```

```python
report = {'name':("Faizal", 'Abhisehek','Nigga'), 'Marks':(90,90,90,)}
report.popitem()
report
```

**LOOPS**
- For loop
- While loop
- Nested loop


```python
# for loop is
for i in "Hello":
  print(i)
```

```python
for i in '*':
  for j in range(1,5):
    print(i*j)
```

```python
for i in '&':
  for j in range(1,9):
    print(i*j)
```

```python
for i in range (0,5):
  for j in '@':
    print(i*j)
```

```python
for i in range(4):
  for j in range(i+1):
    print('A', end = ' ')
  print()
```

```python
for i in range(5):
  for j in range(1,i+1):
    print(i, end=' ')
  print()
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue

  print(x)
```

```python
lst=[23,55,6,8,13,11,90,22,97,64,28]
even=[]
odd=[]
for i in lst:
  if i %2==0:
    even.append(i)
  else:
    odd.append(i)
print(even)
print(odd)
```
